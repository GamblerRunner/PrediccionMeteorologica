```python
from pyspark.sql import SparkSession
from pyspark.sql.window import Window
import pyspark.sql.functions as F
import pandas as pd
from pyspark.sql.functions import col, sin, cos, lit
import math
from pyspark.sql.functions import pandas_udf 
import itertools
from statsmodels.tsa.statespace.sarimax import SARIMAX
from statsmodels.tsa.arima.model import ARIMA
from pyspark.sql.types import StructType, StructField, DateType, DoubleType
import matplotlib.pyplot as plt
import numpy as np
from pyspark.ml.feature import VectorAssembler
from pyspark.ml.regression import RandomForestRegressor
from pyspark.ml.evaluation import RegressionEvaluator
from datetime import datetime, timedelta
```

```python
spark = SparkSession.builder.appName("SeriesTemporales").getOrCreate()
# \PrediccionMeteorologica\data\homogenized\diarios\indicativo=ESMAD2800000028240A\tipo_dato=humedad_media
df = spark.read.parquet("../../../data/transformed/diarios/indicativo=ESMAD2800000028240A/tipo_dato=humedad_media",header=True, inferSchema=True)
df = df.withColumn("fecha", F.to_date("fecha"))

df = df.orderBy(["fecha"])
# Mostrar datos
#df.groupBy("anio").count().show()
#df = df.withColumn("date", F.to_date("date"))  
df.show()
```

```python
window = Window.orderBy("fecha")
df = df.withColumn("lag_7", F.lag("value", 7).over(window)).na.drop()

# 4. Preparar modelo
assembler = VectorAssembler(
    inputCols=["lag_7", "sin_dia", "cos_dia", "sin_mes", "cos_mes"],
    outputCol="features"
)

df_assembled = assembler.transform(df)

max_date = df_assembled.agg({"fecha": "max"}).collect()[0][0]
split_date = max_date - timedelta(days=7)

test = df_assembled.filter(F.col("fecha") > split_date)
train = df_assembled.filter(F.col("fecha") <= split_date)

rf = RandomForestRegressor(featuresCol="features", labelCol="value")
model = rf.fit(train)
predictions = model.transform(test).withColumnRenamed("prediction", "humedad")

```

```python
pdf = predictions.select("fecha", "value", "humedad").toPandas()
pdf.to_excel("humedad_ultimos_7_dias.xlsx", index=False)
```

```python
plt.figure(figsize=(10, 5))
plt.plot(pdf["fecha"], pdf["value"], "bo-", label="Real")
plt.plot(pdf["fecha"], pdf["humedad"], "rs--", label="Pronóstico")
plt.title("Humedad - Últimos 7 días")
plt.xlabel("Fecha")
plt.ylabel("Valor")
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
