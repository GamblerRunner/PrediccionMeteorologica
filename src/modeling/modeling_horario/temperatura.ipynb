```python
from pyspark.sql import SparkSession
from pyspark.sql.window import Window
import pyspark.sql.functions as F
import pandas as pd
from pyspark.sql.functions import col, sin, cos, lit
import math
from pyspark.sql.functions import pandas_udf 
import itertools
from statsmodels.tsa.statespace.sarimax import SARIMAX
from statsmodels.tsa.arima.model import ARIMA
from pyspark.sql.types import StructType, StructField, DateType, DoubleType
import matplotlib.pyplot as plt
import numpy as np
from pyspark.ml.feature import VectorAssembler
from pyspark.ml.regression import RandomForestRegressor
from pyspark.ml.evaluation import RegressionEvaluator
from datetime import datetime, timedelta
from pyspark.sql import SparkSession
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.statespace.sarimax import SARIMAX
from pathlib import Path
```

```python
spark = SparkSession.builder.appName("OpenMeteoDataProcessing").getOrCreate()

# Definir la ruta base
data_path = Path("../../../data/raw/Open_Meteo/diarios/Hoyo de Manzanares_data")
print("Ruta absoluta:", data_path.resolve())

# Lista para almacenar DataFrames individuales
dfs = []

# Recorrer los directorios por año
for entry in data_path.glob("*"):
    if entry.is_dir() and entry.name.isdigit():  # Filtra solo directorios con nombres numéricos (años)
        year = int(entry.name)
        print(f"[DEBUG] Procesando año: {year}")
        if year > 2023: 
            # Iterar sobre los meses dentro del año
            for month_dir in entry.glob("*"):
                if month_dir.is_dir():
                    print(f"[DEBUG] Procesando mes: {month_dir.name}")
                    
                    # Leer todos los CSV del mes
                    for csv_file in month_dir.glob("*.csv"):
                        try:
                            df = spark.read.option("header", True).option("inferSchema", True).csv(str(csv_file))
                            if "date" in df.columns:
                                df = df.withColumn("date", col("date"))  # Asegurar que la columna date existe
                                dfs.append(df)
                            else:
                                print(f"[ERROR] Archivo {csv_file} no tiene la columna 'date'")
                        except Exception as e:
                            print(f"[ERROR] Fallo al leer {csv_file}: {e}")

if dfs:
    final_df = dfs[0]
    for df in dfs[1:]:
        final_df = final_df.union(df)
    
    print("[INFO] Datos combinados correctamente")
    final_df.show()
else:
    print("[ERROR] No se cargaron datos. Verifica la estructura de carpetas y los archivos CSV.")
```

```python
final_df = final_df.withColumn("date", F.to_timestamp("date")) \
                      .filter(F.year("date") >= 2024)
final_df.show()
```

```python
window = Window.orderBy("date")

final_df = final_df.withColumn("lag24_temp", F.lag("temperature_2m", 24).over(window)) \
                   .withColumn("lag24_humidity", F.lag("relative_humidity_2m", 24).over(window)) \
                   .na.drop()

# Características cíclicas horarias
final_df = final_df.withColumn("hour", F.hour("date")) \
                   .withColumn("sin_hora", F.sin(2 * np.pi * F.col("hour") / 24)) \
                   .withColumn("cos_hora", F.cos(2 * np.pi * F.col("hour") / 24))

# Preparar modelo
assembler = VectorAssembler(
    inputCols=["lag24_temp", "lag24_humidity", "sin_hora", "cos_hora"],
    outputCol="features"
)
df_assembled = assembler.transform(final_df)

# Generar próximas 24 horas
last_date = df_assembled.agg(F.max("date")).first()[0]
future_dates = [(last_date + timedelta(hours=i),) for i in range(1, 25)]
future_df = spark.createDataFrame(future_dates, ["date"])

# Unir datos históricos + futuros y calcular lags
full_df = df_assembled.unionByName(future_df, allowMissingColumns=True)
full_df = full_df.withColumn("lag24_temp", F.lag("temperature_2m", 24).over(window)) \
                 .withColumn("lag24_humidity", F.lag("relative_humidity_2m", 24).over(window)) \
                 .withColumn("hour", F.hour("date")) \
                 .withColumn("sin_hora", F.sin(2 * np.pi * F.col("hour") / 24)) \
                 .withColumn("cos_hora", F.cos(2 * np.pi * F.col("hour") / 24))

# Filtrar solo las 24 horas a predecir
future_data = full_df.filter(F.col("date") > last_date)

# Entrenar y predecir
rf = RandomForestRegressor(featuresCol="features", labelCol="temperature_2m")
model = rf.fit(df_assembled)
predictions = model.transform(future_data)

# Mostrar predicciones
predictions.select("date", "prediction").withColumnRenamed("prediction", "temp_predicha").show(24)
```
